```python
import pandas as pd
import missingno as msno
import seaborn as sns
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
listings_df = pd.read_csv('./datasets/listings.csv')
listings_df
```

```python
msno.bar(listings_df)
```

```python
#null content of listing.csv:

#has_availability: all null. Don't need
#bathrooms
#bedrooms
#bed
#name
#review_scores_checkin
#review_scores_cleanliness
#review_scores_communication
#review_scores_location
#review_scores_rating
#review_scores_value
#weekly_price
#zipcode
```

```python
#dropping has_availability because that column is useless
listings_df = listings_df.drop(['has_availability'], axis=1)
```

```python
msno.matrix(listings_df)
```

```python
#bathrooms, bedrooms, and beds have nans but are very few. The reasoning for them missing is probably fairly random.
```

```python
sns.distplot(listings_df.bathrooms)
```

```python
#bathrooms seems fairly skewed. So we will fill nan with median:
listings_df['bathrooms'] = listings_df['bathrooms'].fillna(listings_df['bathrooms'].median())
```

```python
sns.distplot(listings_df.bedrooms)
```

```python
#bedrooms seems fairly skewed. So we will fill nan with median:
listings_df['bedrooms'] = listings_df['bedrooms'].fillna(listings_df['bedrooms'].median())
```

```python
sns.distplot(listings_df.beds)
```

```python
#bedrooms seems fairly skewed. So we will fill nan with median:
listings_df['beds'] = listings_df['beds'].fillna(listings_df['beds'].median())
```

```python
listings_df['name'] = listings_df['name'].fillna('missing')
```

```python
sns.distplot(listings_df.review_scores_checkin)
```

```python
#bedrooms seems fairly skewed. So we will fill nan with median:
listings_df['review_scores_checkin'] = listings_df['review_scores_checkin'].fillna(listings_df['review_scores_checkin'].median())
```

```python
listings_df['review_scores_cleanliness'] = listings_df['review_scores_cleanliness'].fillna(listings_df['review_scores_cleanliness'].median())
```

```python
listings_df['review_scores_communication'] = listings_df['review_scores_communication'].fillna(listings_df['review_scores_communication'].median())
```

```python
listings_df['review_scores_location'] = listings_df['review_scores_location'].fillna(listings_df['review_scores_location'].median())
listings_df['review_scores_rating'] = listings_df['review_scores_rating'].fillna(listings_df['review_scores_rating'].median())
listings_df['review_scores_value'] = listings_df['review_scores_value'].fillna(listings_df['review_scores_value'].median())
```

```python
#converts price from string to float
for ii in range(len(listings_df)):
    listings_df['price'][ii] = float(listings_df['price'][ii][1:].replace(',', ''))
```

```python
listings_df['weekly_price'] = listings_df.apply(
    lambda row: row['price']*7 if pd.isnull(row['weekly_price']) else float(row['weekly_price'][1:].replace(',', '')),
    axis=1
)
```

```python
#converts zipcode
def convert_to_zipcode(code):
    if code[:5] == 'Near ':
        return str(int(float(code[5:])))
    if code.endswith('.0'):
        return str(int(float(code[:-2])))
    if len(str(code)) > 5:
        return code
    code = str(int(float(code)))
    while len(code) < 5:
        code = '0' + code
    return code

#zipcode must be 5 digits and if it is less than 5 digits, we append 0 to it?
#if there is NaN, we replace it with missing
listings_df['zipcode'] = listings_df.apply(
    lambda row: 'missng' if pd.isnull(row['zipcode']) else convert_to_zipcode(row['zipcode']),
    axis=1
)
```

```python
def amenities_to_lst(amenities):
    return amenities[1:-1].split(',')

listings_df['amenities'] = listings_df.apply(lambda row: amenities_to_lst(row['amenities']), axis=1)

listings_df.amenities
```

```python
res = pd.get_dummies(pd.DataFrame(listings_df.amenities.values.tolist()),prefix_sep ='',prefix='').sum(level=0, axis = 1)
```
